```python
import numpy as np
```

```python
#Примеры создания массивов
```

```python
a = np.array([3, 4, 5, 6]) #создание одномерного массива
a
```

```python
c = np.arange(0, 13, 4) #создание массива от 0 до 12 с шагом 4
c
```

```python
b = np.array([[c], [c + 4]]) #создание двумерного массива
b
```

```python
b = np.ones([4, 5]) #создание двумерного массива из единиц
b
```

```python
b = np.zeros([5, 4]) #создание двумерного массива из нулей
b
```

```python
b = np.eye(4) #создание единичной матрицы
b 
```

```python
b = np.random.uniform(1, 11, (5, 5)) #создание двумерного массива с равномерно распределёнными случайными числами
b
```

```python
#Демонстрация атрибутов
```

```python
np.set_printoptions(precision=2) #отбрасывание дробной части до стоых при выводе
print(f'Массив: \n {b}')
print(f'Размеры массива: {b.shape}')
print(f'Тип данных: {b.dtype}')
```

```python
#Примеры индексирования и срезов
```

```python
b[2][2] #элемент в центре матрицы
```

```python
b[:3] #первые три строки
```

```python
b[::,1:4] #три столбца в середине массива
```

```python
#Примеры векторных операций и базовых статистик
```

```python
a * 2 #поэлементное умножение массива
```

```python
a - 2 #поэлементное вычитание массива
```

```python
a + c #поэлементное сложение двух массивов
```

```python
print(f'Минимальное: {np.min(a + c)}')
print(f'Максимальное: {np.max(a + c)}')
print(f'Среднее: {np.mean(a + c)}')
```
